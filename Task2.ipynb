```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Load the dataset
data = pd.read_csv('iris.csv')  # Replace with the correct path to your Iris dataset

# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(data.head())

# Data Cleaning

# Check for missing values
print("\nMissing values in the dataset:")
print(data.isnull().sum())

# Impute or remove missing values if necessary (assuming there are no missing values in Iris dataset)
# data.fillna(method='ffill', inplace=True)  # Forward fill for demonstration

# Encode categorical variables
label_encoder = LabelEncoder()
data['species'] = label_encoder.fit_transform(data['species'])

# Standardize the data (only numerical columns)
scaler = StandardScaler()
numerical_columns = data.columns[:-1]  # All columns except 'species'
data[numerical_columns] = scaler.fit_transform(data[numerical_columns])

# Display the first few rows after preprocessing
print("\nFirst few rows after preprocessing:")
print(data.head())

# Exploratory Data Analysis (EDA)

# Distribution of each feature
plt.figure(figsize=(16, 12))
plt.suptitle('Distribution of Features', fontsize=20)
for i, column in enumerate(data.columns[:-1], 1):  # Exclude 'species' column
    plt.subplot(2, 2, i)
    sns.histplot(data[column], kde=True, bins=20, color='skyblue', edgecolor='black')
    plt.xlabel(column)
    plt.ylabel('Frequency')
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()

# Pairplot to visualize relationships between variables
sns.pairplot(data, hue='species', markers=["o", "s", "D"])
plt.suptitle('Pairplot of Features', fontsize=20)
plt.show()

# Correlation matrix
# Only use numerical columns for the correlation matrix
numerical_data = data[numerical_columns]
plt.figure(figsize=(10, 8))
sns.heatmap(numerical_data.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix', fontsize=20)
plt.show()

# Boxplot for each feature grouped by species
plt.figure(figsize=(16, 12))
plt.suptitle('Boxplot of Features by Species', fontsize=20)
for i, column in enumerate(data.columns[:-1], 1):  # Exclude 'species' column
    plt.subplot(2, 2, i)
    sns.boxplot(x='species', y=column, data=data)
    plt.xlabel('Species')
    plt.ylabel(column)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
